```python
vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(3)
```

```python
def run_score(batsman_name):
    vk=delivery[delivery['batsman']==batsman_name]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).index[0]
```

```python
run_score('V Kohli')
```

## The isin() function

Find the most destructive death over batsman in the history of IPL
on the basis of strike rate = (number of runs/Number of Balls)/100
Min batsman 200 balls in over 16-20

```python
death_over=delivery[delivery['over']>15]
```

```python
all_batsman=death_over.groupby('batsman')['batsman_runs'].count()
x=all_batsman>200
batsman_list=all_batsman[x].index.tolist()
```

```python
# Runs scored by all these 43 batsman
# Balls playes by these 43 batsman

final=delivery[delivery['batsman'].isin(batsman_list)]
```

```python
run=final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls=final.groupby('batsman')['batsman_runs'].count()
```

```python
sr=(run/balls)*100
```
